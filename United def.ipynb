```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
MUDef = pd.read_csv("FBRef Data/Manchester United/ManUtd Defensive Actions.csv", encoding="latin1")
MUGoal = pd.read_csv("FBRef Data/Manchester United/ManUtd Goal and Shot Creation.csv", encoding="latin1")
MUMisc = pd.read_csv("FBRef Data/Manchester United/ManUtd Misc.csv", encoding="latin1")
MUPType = pd.read_csv("FBRef Data/Manchester United/ManUtd Pass Type.csv", encoding="latin1")
MUPoss = pd.read_csv("FBRef Data/Manchester United/ManUtd Possesion.csv", encoding="latin1")
MUShoot =pd.read_csv("FBRef Data/Manchester United/ManUtd Shooting.csv", encoding="latin1")
MUStd =pd.read_csv("FBRef Data/Manchester United/ManUtd Standard.csv", encoding="latin1")
```

```python
MUGoal.drop(columns=["Matches", "Nation"], inplace=True)
MUDef.drop(columns=["Matches", "Nation"], inplace=True)
MUPType.drop(columns=["Matches", "Nation"], inplace=True)
MUPoss.drop(columns=["Matches", "Nation"], inplace=True)
MUShoot.drop(columns=["Matches", "Nation"], inplace=True)
MUStd.drop(columns=["Matches", "Nation"], inplace=True)
MUMisc.drop(columns=["Matches", "Nation"], inplace=True)
```

```python
all_def = pd.read_csv("FBRef Data/Defensive Action.csv", encoding="latin1")
all_ptype = pd.read_csv("FBRef Data/Passing Stats.csv", encoding="latin1")
all_pass = pd.read_csv("FBRef Data/Pass.csv", encoding="latin1")
all_poss = pd.read_csv("FBRef Data/Possesion.csv", encoding="latin1")
all_misc = pd.read_csv("FBRef Data/Misc.csv", encoding="latin1")

```

```python
prem_def = all_def[all_def["Comp"]=='eng\xa0Premier League']
prem_ptype = all_ptype[all_ptype["Comp"]=='eng\xa0Premier League']
prem_pass = all_pass[all_pass["Comp"]=='eng\xa0Premier League']
prem_poss = all_poss[all_poss["Comp"]=='eng\xa0Premier League']
prem_misc = all_misc[all_misc["Comp"]=='eng\xa0Premier League']
```

```python
def_prem_def = prem_def[prem_def["Pos"]=='DF']
def_prem_ptype = prem_ptype[prem_ptype["Pos"]=='DF']
def_prem_pass = prem_pass[prem_pass["Pos"]=='DF']
def_prem_poss = prem_poss[prem_poss["Pos"]=='DF']
def_prem_misc = prem_misc[prem_misc["Pos"]=='DF']
```

```python
def_prem_misc["90s"] = def_prem_misc["90s"].astype("float64")
def_prem_poss["90s"] = def_prem_poss["90s"].astype("float64")
```

```python
def_prem_def = def_prem_def[def_prem_def["90s"] > 18.0]
def_prem_ptype = def_prem_ptype[def_prem_ptype["90s"] > 18.0]
def_prem_pass = def_prem_pass[def_prem_pass["90s"] > 18.0]
def_prem_poss = def_prem_poss[def_prem_poss["90s"] > 18.0]
def_prem_misc = def_prem_misc[def_prem_misc["90s"] > 18.0]
```

```python
def_prem_def.reset_index(inplace=True)
def_prem_ptype.reset_index(inplace=True)
def_prem_pass.reset_index(inplace=True)
def_prem_poss.reset_index(inplace=True)
def_prem_misc.reset_index(inplace=True)
```

```python
def_prem_def.drop(columns=["index" ,"Rk", "Nation", "Matches","Pos" ,"Comp" ,"Age" ,"Born"], inplace=True)
def_prem_ptype.drop(columns=["index" ,"Rk", "Nation", "Matches","Pos" ,"Comp" ,"Age" ,"Born"], inplace=True)
def_prem_pass.drop(columns=["index" ,"Rk", "Nation", "Matches","Pos" ,"Comp" ,"Age" ,"Born"], inplace=True)
def_prem_poss.drop(columns=["index" ,"Rk", "Nation", "Matches","Pos" ,"Comp" ,"Age" ,"Born"], inplace=True)
def_prem_misc.drop(columns=["index" ,"Rk", "Nation", "Matches","Pos" ,"Comp" ,"Age" ,"Born"], inplace=True)
```
