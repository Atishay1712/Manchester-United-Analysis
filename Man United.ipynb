```python
all_def = pd.read_csv("FBRef Data/Defensive Action.csv", encoding="latin1")
```

```python
all_def["Comp"][0]
```

```python
prem_def = all_def[all_def["Comp"]=='eng\xa0Premier League']
```

```python
prem_defact_mids = prem_def[(prem_def["Pos"]== "MF") ]
```

```python
prem_defact_mids = prem_defact_mids[prem_defact_mids["90s"] > 18.0]

prem_defact_mids = prem_defact_mids.sort_values("TklW", ascending=False)
prem_defact_mids.reset_index(inplace=True)
prem_defact_mids.drop(columns=["index" ,"Rk", "Nation", "Matches","Pos" ,"Comp" ,"Age" ,"Born"], inplace=True)
```

```python
prem_defact_mids.columns
```

```python
x = prem_defact_mids["TklW"]
y = prem_defact_mids["Succ"]
z = prem_defact_mids["Player"]
fig,ax = plt.subplots(1,1,figsize=(7,7))
for i in range(20):
    if prem_defact_mids["Squad"][i] == "Manchester Utd":
        ax.scatter(x[i],y[i],c ="red")
    else:
        ax.scatter(x[i],y[i],c="blue")
ax.set_xlabel("Tackles per 90")
ax.set_ylabel("Pressure per 90")
plt.title("Tackles v Pressure per 90")
for i in range(20):
    ax.annotate(z.iloc[i],(x.iloc[i]+.01,y.iloc[i]+0.01))

    
plt.savefig("tkl v press for prem mids")
```

```python
prem_defact_mids = prem_defact_mids.sort_values("Blocks", ascending=False)
prem_defact_mids.reset_index(inplace=True)
x = prem_defact_mids["Blocks"]
y = prem_defact_mids["Int"]
z = prem_defact_mids["Player"]
fig,ax = plt.subplots(1,1,figsize=(7,7))
for i in range(20):
    if prem_defact_mids["Squad"][i] == "Manchester Utd":
        ax.scatter(x[i],y[i],c ="red")
    else:
        ax.scatter(x[i],y[i],c="blue")
ax.set_xlabel("Blocks per 90")
ax.set_ylabel("Int per 90")
plt.title("Blocks v Int per 90")
for i in range(20):
    ax.annotate(z.iloc[i],(x.iloc[i]+.01,y.iloc[i]+0.01))

    
plt.savefig("Blocks v Int for prem mids")
```
